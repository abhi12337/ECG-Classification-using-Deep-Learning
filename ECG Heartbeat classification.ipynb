```python
def plot_equilibre(equilibre):
    plt.figure(figsize=(10,10))
    my_circle=plt.Circle( (0,0), 0.7, color='white')
    plt.pie(equilibre, labels=['n','q','v','s','f'], colors=['red','green','blue','skyblue','orange'],autopct='%1.1f%%')
    p=plt.gcf()
    p.gca().add_artist(my_circle)
    plt.show()
```

```python
X_train = df_train.values[:, :-1]
y_train = df_train.values[:, -1].astype(int)

X_test  = df_test.values[:, :-1]
y_test  = df_test.values[:, -1].astype(int)

```

```python
C0 = np.argwhere(y_train == 0).flatten()
C1 = np.argwhere(y_train == 1).flatten()
C2 = np.argwhere(y_train == 2).flatten()
C3 = np.argwhere(y_train == 3).flatten()
C4 = np.argwhere(y_train == 4).flatten()

x = np.arange(0, 187)*8/1000
plt.figure(figsize=(20,6))
plt.plot(x, X_train[C0, :][0], label="Normal")
plt.plot(x, X_train[C1, :][0], label="Artial Premature")
plt.plot(x, X_train[C2, :][0], label="Premature ventricular contraction")
plt.plot(x, X_train[C3, :][0], label="Fusion of ventricular and normal")
plt.plot(x, X_train[C4, :][0], label="Fusion of paced and normal")
plt.legend()
plt.title("1-beat ECG for every category", fontsize=20)
plt.ylabel("Amplitude", fontsize=15)
plt.xlabel("Time (ms)", fontsize=15)
plt.show()
```

```python
X_train = np.expand_dims(X_train, 2)
X_test = np.expand_dims(X_test, 2)

print(X_train.shape)
print(X_test.shape)
```

```python
from tensorflow.keras import utils
y_train = utils.to_categorical(y_train)
y_test  = utils.to_categorical(y_test)

print(y_train.shape)
print(y_test.shape)
```

```python
n_obs, feature, depth = X_train.shape
num_classes= 5
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv1D, MaxPooling1D, Flatten, Dense, Activation, Add

# Input layer
inputs = Input(shape=(feature, depth))

# Layer 1
conv1 = Conv1D(filters=16, kernel_size=3, strides=1, padding='valid')(inputs)
conv1 = Activation('relu')(conv1)

# Layer 2
conv2 = Conv1D(filters=16, kernel_size=3, strides=1, padding='valid')(conv1)
conv2 = Activation('relu')(conv2)

# Layer 3
maxpool1 = MaxPooling1D(pool_size=2, strides=2)(conv2)

# Layer 4
conv3 = Conv1D(filters=32, kernel_size=3, strides=1, padding='valid')(maxpool1)
conv3 = Activation('relu')(conv3)

# Layer 5
conv4 = Conv1D(filters=32, kernel_size=3, strides=1, padding='valid')(conv3)
conv4 = Activation('relu')(conv4)

# Layer 6
maxpool2 = MaxPooling1D(pool_size=2, strides=2)(conv4)

# Layer 7
conv5 = Conv1D(filters=64, kernel_size=3, strides=1, padding='valid')(maxpool2)
conv5 = Activation('relu')(conv5)

# Layer 8
conv6 = Conv1D(filters=64, kernel_size=3, strides=1, padding='valid')(conv5)
conv6 = Activation('relu')(conv6)

# Layer 9
maxpool3 = MaxPooling1D(pool_size=2, strides=2)(conv6)

# Layer 10
conv7 = Conv1D(filters=128, kernel_size=3, strides=1, padding='valid')(maxpool3)
conv7 = Activation('relu')(conv7)

# Layer 11
conv8 = Conv1D(filters=128, kernel_size=3, strides=1, padding='valid')(conv7)
conv8 = Activation('relu')(conv8)

# Layer 12
conv9 = Conv1D(filters=128, kernel_size=3, strides=1, padding='valid')(conv8)
conv9 = Activation('relu')(conv9)

# Layer 13
maxpool4 = MaxPooling1D(pool_size=2, strides=2)(conv9)

# Layer 14
flatten = Flatten()(maxpool4)

# Layer 15
fc1 = Dense(units=30)(flatten)
fc1 = Activation('relu')(fc1)

# Layer 16
output = Dense(units=5)(fc1)
output = Activation('softmax')(output)

# Create the model
model = Model(inputs=inputs, outputs=output)

model.summary()
```

```python
from tensorflow.keras.optimizers import Adam
model.compile(loss='categorical_crossentropy', optimizer=Adam(1e-4), metrics=['accuracy'])
```

```python
batch_size = 100
num_epochs = 30
```

```python
history = model.fit(X_train, y_train, epochs=num_epochs, batch_size=batch_size, validation_data=(X_test, y_test))
```

```python
history.history['loss']
```

```python
plt.plot(np.arange(1, 1+len(history.history['loss'])), history.history['loss'], label = 'Training Loss')
plt.plot(np.arange(1, 1+len(history.history['val_loss'])), history.history['val_loss'], label = 'Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
```

```python
model.save('ecg_arrhythmia.hdf5')
```

```python
from sklearn.metrics import classification_report, confusion_matrix

labels = ["Normal",
          "Artial Premature",
          "Premature ventricular contraction",
          "Fusion of ventricular and normal",
          "Fusion of paced and normal"]
```

```python
predY = model.predict(X_test)
y_pred = np.argmax(predY,axis=1)
y_actual = np.argmax(y_test,axis=1)
cm = confusion_matrix(y_actual, y_pred)
print(cm)
```

```python
model.evaluate(X_test,y_test)
```

```python
print(classification_report(y_actual, y_pred, target_names=labels))
```
